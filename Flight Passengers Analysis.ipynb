# Flight Data Analysis

```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the flights dataset
flights = sns.load_dataset('flights')
flights
```

```python
# Display the first few rows of the dataset
print("First 5 rows of the dataset:")
print(flights.head())
```

```python
flights.info()
```

```python
# Summary statistics
print("\nSummary Statistics:")
print(flights.describe())

```

```python
# Check for missing values
print("\nMissing Values:")
print(flights.isnull().sum())

```

```python
# Bar plot of passengers by month
plt.figure(figsize=(10, 6))
sns.barplot(x='month', y='passengers', data=flights, ci=None)
plt.xlabel('Month')
plt.ylabel('Passengers')
plt.title('Total Passengers by Month')
plt.show()

```

```python
# Line plot of passengers over time
plt.figure(figsize=(10, 6))
sns.lineplot(x='year', y='passengers', data=flights)
plt.xlabel('Year')
plt.ylabel('Passengers')
plt.title('Passenger Trend Over Time')
plt.show()

```

```python
# Box plot of passengers by month
plt.figure(figsize=(10, 6))
sns.boxplot(x='month', y='passengers', data=flights)
plt.xlabel('Month')
plt.ylabel('Passengers')
plt.title('Passenger Distribution by Month')
plt.show()

```

```python
# Violin plot of passengers by year
plt.figure(figsize=(10, 6))
sns.violinplot(x='year', y='passengers', data=flights)
plt.xlabel('Year')
plt.ylabel('Passengers')
plt.title('Passenger Distribution by Year')
plt.show()
```

```python
# Heatmap of passenger counts
pivot_table = flights.pivot_table(index='month', columns='year', values='passengers', aggfunc=np.sum)
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, cmap='Greens', annot=True, fmt='d', linewidths=0.5)
plt.xlabel('Year')
plt.ylabel('Month')
plt.title('Heatmap of Passenger Counts by Month and Year')
plt.show()
```

```python
# Swarm plot of passengers by month
plt.figure(figsize=(10, 6))
sns.swarmplot(x='month', y='passengers', data=flights)
plt.xlabel('Month')
plt.ylabel('Passengers')
plt.title('Swarm Plot of Passenger Counts by Month')
plt.show()
```

```python
# ECDF plot of passenger counts
plt.figure(figsize=(10, 6))
sns.ecdfplot(data=flights, x='passengers', hue='month')
plt.xlabel('Passengers')
plt.ylabel('ECDF')
plt.title('ECDF Plot of Passenger Counts by Month')
plt.legend(title='Month')
plt.show()
```

```python
# KDE plot of passenger counts
plt.figure(figsize=(10, 6))
sns.kdeplot(data=flights['passengers'], shade=True)
plt.xlabel('Passengers')
plt.ylabel('Density')
plt.title('Kernel Density Estimation (KDE) Plot of Passenger Counts')
plt.show()
```

```python
# Pie chart of total passengers by year
plt.figure(figsize=(8, 6))
yearly_passengers = flights.groupby('year')['passengers'].sum()
plt.pie(yearly_passengers, labels=yearly_passengers.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Proportion of Total Passengers by Year')
plt.axis('equal')
plt.show()
```

```python
# Pairplot of passengers by month and year
plt.figure(figsize=(10, 6))
sns.pairplot(flights, hue='month', markers='o')
plt.suptitle('Pairplot of Passenger Count by Month and Year', y=1.02)
plt.show()
```
