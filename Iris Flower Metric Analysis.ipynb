# Iris Data Analysis

```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the Iris dataset
iris = sns.load_dataset('iris')
iris
```

```python
# Encode the 'species' column into numerical values
label_encoder = LabelEncoder()
iris['species_encoded'] = label_encoder.fit_transform(iris['species'])
```

```python
# Drop the original 'species' column
iris.drop('species', axis=1, inplace=True)
iris
```

```python
# Display the first few rows of the dataset
print("First 5 rows of the dataset:")
print(iris.head())
```

```python
# Summary statistics
print("\nSummary Statistics:")
print(iris.describe())
```

```python
# Check for missing values
print("\nMissing Values:")
print(iris.isnull().sum())
```

```python
# Pairplot to visualize relationships between variables
sns.pairplot(iris, hue='species_encoded', markers=["o", "s", "D"])
plt.show()
```

```python
# Boxplot to visualize distribution of each feature
plt.figure(figsize=(10,6))
plt.title('Boxplot of Iris Features')
sns.boxplot(data=iris)
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(8,6))
plt.title('Correlation Heatmap')
sns.heatmap(iris.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.show()
```

```python
# EDA
df = sns.load_dataset('iris')
df
```

# Diiferent types of Charts

# 1. Bar Chart

```python
# Bar plot of average sepal length by species
plt.figure(figsize=(8, 5))
sns.barplot(x='species', y='sepal_length', data=df)
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.title('Average Sepal Length by Species')
plt.show()
```

# 2. Line Chart

```python
# Line chart
plt.figure(figsize=(8, 6))
sns.lineplot(x=range(len(iris)), y='petal_width', data=iris)
plt.xlabel('Index')
plt.ylabel('Petal Width')
plt.title('Petal Width Over Index')
plt.show()
```

# 3. Pie Chart

```python
# Pie chart
plt.figure(figsize=(8, 6))
iris['species'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Distribution of Iris Species')
plt.ylabel('')
plt.show()
```

# 4. Histogram

```python
# Histogram
plt.figure(figsize=(8, 6))
sns.histplot(data=iris, x='sepal_width', bins=20, kde=True)
plt.xlabel('Sepal Width')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Width')
plt.show()
```

# 5. Box Plot

```python
# Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='sepal_length', data=iris, hue = 'species')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.title('Boxplot of Sepal Length by Species')
plt.show()
```

# 6. Violin Plot

```python
# Violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(x='species', y='sepal_width', data=iris)
plt.xlabel('Species')
plt.ylabel('Sepal Width')
plt.title('Violin Plot of Sepal Width by Species')
plt.show()
```

# 7. Scatter Plot

```python
# Scatter plot matrix
plt.figure(figsize=(10, 8))
sns.pairplot(iris, hue='species')
plt.suptitle('Scatter Plot Matrix of Iris Dataset', y=1.02)
plt.show()
```

# 8. Swarm Plot 

```python
# Swarm plot
plt.figure(figsize=(8, 6))
sns.swarmplot(x='species', y='petal_width', data=iris)
plt.xlabel('Species')
plt.ylabel('Petal Width')
plt.title('Swarm Plot of Petal Width by Species')
plt.show()
```

# 9. PairGrid

```python
# PairGrid
g = sns.PairGrid(iris)
g.map_upper(sns.scatterplot)
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
plt.suptitle('PairGrid of Iris Dataset', y=1.02)
plt.show()
```

# 10. Joint Plot

```python
# Joint plot
plt.figure(figsize=(8, 6))
sns.jointplot(x='sepal_length', y='sepal_width', data=iris, kind='hex')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Joint Plot of Sepal Length vs. Sepal Width')
plt.show()
```

# 11. Cluster Map

```python
# Clustermap
plt.figure(figsize=(8, 6))
sns.clustermap(iris.drop(columns='species').corr(), cmap='coolwarm', annot=True)
plt.title('Clustermap of Iris Dataset')
plt.show()
```

# 12. KDE Plot

```python
# KDE plot
plt.figure(figsize=(8, 6))
sns.kdeplot(data=iris['sepal_length'], shade=True)
plt.xlabel('Sepal Length')
plt.ylabel('Density')
plt.title('Kernel Density Estimation (KDE) Plot of Sepal Length')
plt.show()
```

# 13. Joint plot with regression line

```python
# Joint plot with regression line
plt.figure(figsize=(8, 6))
sns.jointplot(x='petal_length', y='petal_width', data=iris, kind='reg')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Joint Plot of Petal Length vs. Petal Width with Regression Line')
plt.show()
```

# 14. Strip plot

```python
# Strip plot
plt.figure(figsize=(8, 6))
sns.stripplot(x='species', y='sepal_length', data=iris, jitter=True)
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.title('Strip Plot of Sepal Length by Species')
plt.show()
```

# 15. ECDF (Empirical Cumulative Distribution Function) Plot

```python
# ECDF plot
plt.figure(figsize=(8, 6))
sns.ecdfplot(data=iris, x='petal_length', hue='species')
plt.xlabel('Petal Length')
plt.ylabel('ECDF')
plt.title('ECDF Plot of Petal Length by Species')
plt.show()
```

# 16. Donut Chart

```python
# Calculate values for the donut chart
species_counts = iris['species'].value_counts()
labels = species_counts.index
sizes = species_counts.values

# Create a donut chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'lightcoral'], wedgeprops=dict(width=0.4))
plt.title('Donut Chart of Iris Species')
plt.axis('equal')
plt.show()
```

# 17. Stacked Bar chart

```python
# Sample data
data = {
    'Group': ['A', 'B', 'C'],
    'Category1': [20, 30, 25],
    'Category2': [15, 25, 30]
}

df = pd.DataFrame(data)

# Stacked bar chart
plt.figure(figsize=(8, 6))
plt.bar(df['Group'], df['Category1'], label='Category 1')
plt.bar(df['Group'], df['Category2'], bottom=df['Category1'], label='Category 2')
plt.xlabel('Group')
plt.ylabel('Count')
plt.title('Stacked Bar Chart')
plt.legend()
plt.show()
```
